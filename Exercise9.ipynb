```python
import pandas as pd
import matplotlib.pyplot as plt

path = pd.read_csv('/content/Technological-Products-Sample-Data - Tech Products.csv')

x = path['Date of Release']
y = path['Price (USD)']

plt.figure(figsize=(16, 8))
plt.title("Price In Every Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.xticks(rotation=45)


plt.plot(x, y, marker='o', color='r', linewidth=3, markersize=8,
         markerfacecolor='blue', markeredgecolor='blue', linestyle='dotted', label='Price Trend')


plt.legend(loc = 'lower right')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/Technological-Products-Sample-Data - Tech Products.csv')


x = data['Date of Release']
y1 = data['Brand']
y2 = data['Device']
y3 = data['Model']

plt.figure(figsize=(10, 15))
plt.title("Product Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales (Units)")
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.plot(x, y1, marker='o', color='r', label='Product A', linestyle='-')
plt.plot(x, y2, marker='s', color='g', label='Product B', linestyle='--')
plt.plot(x, y3, marker='^', color='b', label='Product C', linestyle=':')


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/Technological-Products-Sample-Data - Tech Products.csv')


brand_sales = data.groupby('Brand')['Price (USD)'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(brand_sales['Brand'], brand_sales['Price (USD)'], color='skyblue')


plt.title("Total Sales by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Sales (Units)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Technological-Products-Sample-Data - Tech Products.csv')

country_counts = data.groupby('Country of Origin')['Device'].count()

plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')

plt.title('Most Common Country of Origin for Devices')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Devices')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Technological-Products-Sample-Data - Tech Products.csv')

country_counts = data['Country of Origin'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Devices by Country of Origin')

plt.tight_layout()
plt.show()
```
