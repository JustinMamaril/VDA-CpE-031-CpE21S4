```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/heart_2022_no_nans.csv')
df_limited = df.head(50)


grouped = df_limited.groupby(['Sex', 'PhysicalActivities']).size().unstack()


for sex in grouped.index:
    plt.figure()

    if sex in grouped.index:
        grouped.loc[sex].plot(kind='pie', autopct='%1.1f%%', startangle=90)
        plt.title(f'Physical Activity Distribution for {sex} (First 50 Entries)')
        plt.ylabel('')
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/heart_2022_no_nans.csv')
df_limited = df.head(50)

# Count the number of heart attacks by gender
heart_attack_counts = df_limited.groupby('Sex')['HadHeartAttack'].value_counts().unstack().fillna(0)

# Plotting
heart_attack_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Heart Attack Counts by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Heart Attacks')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(title='Heart Attack', labels=['No', 'Yes'])  # Adjust labels based on your data
max_count = heart_attack_counts.values.max()  # Get the maximum count
plt.yticks(range(0, max_count + 1))  # Set y-ticks from 0 to max count
plt.show()
```
