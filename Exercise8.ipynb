```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C"]
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 + [3]*10]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40,30,30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```

```python
import matplotlib.pyplot as plt

path = "/content/drive/MyDrive/Dataset/mamaril.csv"
df = pd.read_csv(path)

values = df.head(10)
df_subset = df[['Name', values]]
plt.figure(figsize=(10, 5))
plt.bar(df_subset['Lane'], df_subset[values])

plt.title("Bar Chart of First 10 Entries")
plt.xlabel("Column Name")
plt.ylabel("Values")

# Show the plot
plt.show()
```
